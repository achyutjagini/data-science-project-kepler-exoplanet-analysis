```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('cumulative-cleaned.csv')
x = data.head()['rowid']
y = data.head()['stellar_eff_temp']
print(np.corrcoef(x, y))
plt.scatter(x, y) 
plt.title('A plot to show the correlation between Row ID and Stellar Effective Temperature')
plt.xlabel('Row ID')
plt.ylabel('Stellar Effective Temperature')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='yellow') #adding the line of best fit
plt.show()
#The below graph shows a positive correlation whose coeffecient is 0.9278
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('cumulative-cleaned.csv')
x = data.tail()['rowid']
y = data.tail()['transit_duration']
print(np.corrcoef(x, y))
plt.scatter(x, y) 
plt.title('A plot to show the correlation between Row ID and Transit Duration')
plt.xlabel('ROW ID')
plt.ylabel('Transit Duration')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='yellow') #adding the line of best fit
plt.show()
#The below graph shows a negative correlation whose coeffecient is -0.6546
```
